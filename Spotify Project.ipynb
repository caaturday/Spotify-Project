```python
# importing libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
# importing spotify data set
df_tracks = pd.read_csv(r'/Users/caaturday/Documents/Projects/Python/Spotify Project/tracks.csv')
df_tracks.head()
```

```python
# checking for null values in the data set
pd.isnull(df_tracks).sum()
```

```python
df_tracks.info()
```

```python
# least popular songs on spotify
sorted_df = df_tracks.sort_values('popularity', ascending = True).head(10)
sorted_df
```

```python
# printing statistics on the songs
df_tracks.describe().transpose()
```

```python
# printing the most popular songs on spotify
most_popular = df_tracks.query('popularity>90', inplace = False).sort_values('popularity', ascending = False)
most_popular
```

```python
df_tracks.set_index('release_date', inplace=True)
df_tracks.index = pd.to_datetime(df_tracks.index)
df_tracks.head()
```

```python
df_tracks[['artists']].iloc[18]
```

```python
# chaning duration time from ms to seconds
df_tracks['duration'] = df_tracks["duration_ms"]#.apply(lambda x: round(x/1000))
df_tracks.drop('duration_ms', inplace=True, axis = 1)
df_tracks.duration.head()
```

```python
# correlation 
corr_df = df_tracks.drop(['key', 'mode', 'explicit'], axis = 1).corr(method = 'pearson')

# plotting
plt.figure(figsize = (14, 6))
heatmap = sns.heatmap(corr_df, annot = True, fmt = '.1g', vmin = -1, vmax = 1, center = 0, cmap = "inferno")
heatmap.set_title('Correlation HeatMap Between Variable')
heatmap.set_xticklabels(heatmap.get_xticklabels(), rotation = 90)
```

## Positive Correlation between loudness and energy

```python
sample_df = df_tracks.sample(int(0.004*len(df_tracks)))
```

```python
print(len(sample_df))
```

```python
plt.figure(figsize=(10, 6))
sns.regplot(data = sample_df, y = 'loudness', x = 'energy', color = 'red').set(title = "Loudness vs Energy Correlation")
```

```python
plt.figure(figsize=(10, 6))
sns.regplot(data = sample_df, y = 'popularity', x = 'acousticness', color = 'blue').set(title = "Popularity vs Accousticness Correlation")
```
